#### Project 1: Exploratory Data Analysis -Gapminder dataset

 The Gapminder dataset provides population data from 1952 to 2007 (at s year inlevals) for several contries around the world. compare the population of the European countries France ,United Kingdom , Italy , Germany and Spain Over this period Using a Line Chart  

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
from matplotlib import colors, markers

data = pd.read_csv("C:\\Users\\CHIRAG\\Documents\\GitHub\\ML_Studycamp\\datasets\\gapminderDataFiveYear.csv")
df=pd.DataFrame(data)

countries=['France','United Kingdom','Italy','Germany','Spain']
df_europe=df[df['country'].isin(countries)]
plt.figure(figsize=(12,7))

colors={'France':'blue','United Kingdom':'red','Italy':'green','Germany':'black','Spain':'orange'}
for country in countries:
    country_data=df_europe[df_europe['country']==country].sort_values('year')
    plt.plot(country_data['year'],country_data['pop'],label=country,color=colors[country],marker='o',markersize=6,linewidth=2.5)
plt.title('population of european countries over time (1952-2007)',fontsize=16)
plt.xlabel('year',fontsize=14)
plt.ylabel('population(In Millions)',fontsize=14)
plt.legend(title='countries',fontsize=12)
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()







```

#### Project 2:Exploratory Data Analysis -Diamonds dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data2 =pd.read_csv('C:\\Users\\CHIRAG\\Documents\\GitHub\\ML_Studycamp\\datasets\\diamonds.csv')
diamond_df=pd.DataFrame(data2)

filtered_diamonds = diamond_df[(diamond_df['clarity'] == 'SI2') & (diamond_df['color'] == 'E')]

cut_colors = {'Fair': 'red', 'Good': 'orange', 'Very Good': 'yellow', 'Premium': 'green', 'Ideal': 'blue'}
colors = filtered_diamonds['cut'].map(cut_colors)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(filtered_diamonds['carat'], filtered_diamonds['price'], c=colors, alpha=0.7, edgecolors='w', s=50)

handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, alpha=0.7) for color in cut_colors.values()]
plt.legend(handles, cut_colors.keys(), title='Cut Quality', fontsize=12, title_fontsize=14)

plt.title('Scatter Plot: Carat vs Price for Diamonds (Clarity: SI2, Color: E)', fontsize=16, fontweight='bold')
plt.xlabel('Carat (Weight)', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
